```python
stocks_close = stocks['Close']
```

```python
stocks_close = pd.DataFrame(stocks_close)
```

```python
stocks_close
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler = MinMaxScaler()
```

```python
stocks_close
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = stocks_close
y = stocks_close['Close']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

```python
Scaler.fit(X_train)
```

```python
X_train = Scaler.transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
y_test.shape
```

```python
X_test.shape
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, Dropout, LSTM
```

```python
model = Sequential()
```

```python
model.add(LSTM(100, return_sequences=True, input_shape = (X_train.shape[1],1)))
Dropout(0.3)
model.add(LSTM(100, return_sequences=False))
Dropout(0.3)
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.summary()
```

```python
model.fit(x = X_train, y=y_train, epochs=100, validation_data = (X_test, y_test), verbose=1, batch_size = 32)
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
y_pred = model.predict(X_test)
```

```python
# Using Recurrent Neural Network
from tensorflow.keras.layers import SimpleRNN
```

```python
model_RNN = Sequential()
```

```python
model_RNN.add(SimpleRNN(100, return_sequences = True, input_shape = (X_train.shape[1],1)))
Dropout(0.3)
model_RNN.add(SimpleRNN(100, return_sequences = False))
Dropout(0.3)
model_RNN.add(Dense(1))

model_RNN.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model_RNN.summary()
```

```python
model_RNN.fit(x=X_train, y=y_train, validation_data = (X_test, y_test), verbose = 1, epochs = 100)
```

```python
model_RNN.history.history
```

```python
model_RNN_loss = pd.DataFrame(model_RNN.history.history)
```

```python
model_RNN_loss
```

```python
model_RNN_loss.plot()
```
